```python
import h5py
import matplotlib.pyplot as plt
```

```python
# Shanghai Part a
gt = []
count = []
for line in open('output.txt', 'r'):
    lines = [i for i in line.split(';')]
    gt.append(float(lines[1]))
    count.append(float(lines[2]))

plt.plot(gt,count,'x')
plt.plot([0, 2000], [0, 2000])
plt.xlabel('ground true')
plt.ylabel('count')
plt.show()
```

```python
base_dir = 'save_file/JHU/'
base_name = 'val_LR' #'val_WD'
#names = ['1e-4.txt', '4e-4.txt', '16e-4.txt', '25e-6.txt'] #For WD
names = ['25e-7.txt'] #For LR

epochs = {name: [] for name in names}; epochs['base'] = []
MAEs = {name: [] for name in names}; MAEs['base'] = []
for name in names:
    full_name = base_dir + base_name + name
    for line in open(full_name, 'r'):
        split = line.split()
        #print(split[1][:-1])
        epochs[name].append(float(split[1][:-1])) #[1] to get epoch from sentence, [:-1] to get rid of ,
        #print(split[3][:-1])
        MAEs[name].append(float(split[3][:-1])) #[3] to get epoch from sentence, [:-1] to get rid of ,
for line in open(base_dir+'val_base.txt', 'r'):
    split = line.split()
    epochs['base'].append(float(split[1][:-1]))
    MAEs['base'].append(float(split[3][:-1]))
        

fig, ax = plt.subplots(1,1,figsize=(10,7))
ax.set_xlim(0,102)
ax.set_ylim(50,100)
for name in names:
    ax.plot(epochs[name], MAEs[name], label=name.split('.')[0])
ax.plot(epochs['base'], MAEs['base'], label='base')
ax.legend()
plt.show()
```
